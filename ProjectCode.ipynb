```python
# Importing Python Libraries 
import os
import tarfile
import urllib 
import pandas as pd
import numpy as np
```

```python
# Importing Data from the website using Python Library
#OR
#Alternatively we can manually donwload it to the directory

down_addr="https://raw.githubusercontent.com/ageron/handson-ml/master/"
path_data=os.path.join("datasets","housing")
total_url=down_addr+"datasets/housing/housing.tgz"


def get_data(downpath=down_addr,pathdata=path_data):
    if not os.path.isdir(path_data):
        os.makedirs(pathdata,exist_ok=True)
        
    tgz_path=os.path.join(path_data,"housing.tgz")
    urllib.request.urlretrieve(total_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=path_data)
    housing_tgz.close()
    
    



```

```python
#Now splitting the Dataset into Test and Training 
def split_test_train(data,test_ratio):
    shuffled_indices=np.random.permutation(len(data))
    test_set_size=int(len(data)*test_ratio)
    train_indices=shuffled_indices[test_set_size:]
    test_indices=shuffled_indices[:test_set_size]
    return data.iloc[train_indices],data.iloc[test_indices]
    
    
```

```python
train_set,test_set=split_test_train(housing,0.2)
print("Data Split Successfully \n Training Set ::",len(train_set),"\n Test Set ::",len(test_set))
```

```python
#To Prevent the Data from getting randomly splitted again and again 
#Using Cyclic Redundancy Check Sum function  

from zlib import crc32

def test_set_check(identifier,test_ratio):
    return crc32(np.int64(identifier)) & 0xffffffff < test_ratio * 2**32


```

```python
#Splitting Data using the CRC Check sum indice
def split_train_test_by_id(data,test_ratio,id_col):
    ids=data[id_col]
    in_test_set=ids.apply(lambda id_:test_set_check(id_,test_ratio))
    return data.loc[~in_test_set], data.loc[in_test_set]
```

```python
#Creating the ID Col for Housing Project
housing_with_id=housing.reset_index()
train_set,test_set=split_train_test_by_id(housing_with_id,0.2,"index")
```

```python
#now Transforming some attribs to add more insight in dataset
from sklearn.base import BaseEstimator, TransformerMixin

room_ix, bedrooms_ix, population_ix, households_ix=[
    list(housing.columns).index(col)
    for col in ("total_rooms", "total_bedrooms", "population", "households")]

class CombAttribAdder(BaseEstimator, TransformerMixin):
    
    def __init__(self,add_bedrooms_per_room=True):
        self.add_bedrooms_per_room=add_bedrooms_per_room
    
    def fit(self,X,y=None):
        return self;
    
    def transform(self,X):
        rooms_per_households=X[:,room_ix]/X[:,households_ix]
        population_per_households=X[:,population_ix]/X[:,households_ix]
        
        if (self.add_bedrooms_per_room):
            bedrooms_per_room=X[:,bedrooms_ix]/X[:,households_ix]
            return np.c_[X,rooms_per_households,population_per_households,bedrooms_per_room]
        else:
            return np.c_[X,rooms_per_households,population_per_households]
        
        attr_adder=CombAttribAdder(add_bedrooms_per_room=False)
        housing_extra_attribs = attr_adder.transform(housing.values)
```

```python
#Adding Transformation Pipelines and Feature Scaling [Standardization]
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline as pipe
from sklearn.impute import SimpleImputer

num_pipeline=pipe(
[
    ('imputer',SimpleImputer(strategy="median")),
    ('attribs_adder',CombAttribAdder()),
    ('std_scaler',StandardScaler())
])

housing_num_tr=num_pipeline.fit_transform(housing_num)
```

```python
#we can simply use Column Transformation for both cat and num cols
from sklearn.compose import ColumnTransformer as ctrans
from sklearn.preprocessing import OneHotEncoder

num_attribs=list(housing_num)
cat_attribs=["ocean_proximity"]

full_pipelines=ctrans([
    ("num",num_pipeline,num_attribs),
    ("cat",OneHotEncoder(),cat_attribs)
])

housing_prepared=full_pipelines.fit_transform(housing)
```
